# F strings

###  Within the f-string (denoted by the 'f' prefix), curly braces {} act as placeholders for variables or expressions. The values of variables are seamlessly inserted into the string, producing a formatted output.

```python
name = "Alice"
age = 30

greeting = f"Hello, {name}! You are {age} years old."
print(greeting)  # Output: Hello, Alice! You are 30 years old.
```

```python
# -strings permit the evaluation of expressions directly within the curly braces.
width = 10
height = 5

area = f"The area of the rectangle is {width * height}."
print(area)
```

```python
item = "apple"
price = 0.99
quantity = 3
# """ allows multiple line f strings
receipt = f"""
Item: {item}
Price: ${price:.1f}
Quantity: {quantity}
Total: ${price * quantity:.1f}
"""

print(receipt)
```

# Arithmetic Expressions

  # Precedence (PEMDAS)
     1) Paranthesis () : (x - 2) * 3 => (x-2) will be evaluated first
     2) Exponent ** : 2 + 4**2 = 18
     3) Unary Minus - : 2 * -y => y will be computed first  
     4) Mul and Div *, / , // , % (equal precedence for these) : 3 - 4 / 2 = 1
     5) Add and Sub +, -
     6) left to right evaluation : 2 * 6 / 4 => 2*6 is evaluated first
     7) ** is evaluated right to left


     # Compound operators
       1) x +=1 ~ x = x+1 (same format for -)
       2) x *=5 ~ x = x*5 (same format for /)
       3) x %=2 ~ x = x%2
