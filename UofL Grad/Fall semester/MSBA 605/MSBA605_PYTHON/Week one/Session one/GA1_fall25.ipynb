### Problem 2 (2 points)

A company charged a fixed platform fee of $100$ for a $900$ service. The service is for 5 users. Write a program to compute the per user cost with two decimal digits.

````python
platform_fee = 100
service_cost = 900
num_users = 5

# Total cost
total_cost = platform_fee + service_cost

# Per user cost
per_user_cost = total_cost / num_users

# Display result with two decimal places
print(f"Per user cost: ${per_user_cost:.2f}")
````

### Problem 4a (1 point)
In the code cell below, write a Python script that creates a tuple named `time_tuple` that represents an _hour_ (integer from 0-23 inclusive), _minute_ (integer from 0-59 inclusive), and _second_ (integer from 0-59 inclusive) in its 3 elements. Set it for the time 9:16:01 AM. Use the tuple's elements to calculate the number of seconds since midnight, accessing each value using its index `[0]`, `[1]`, [`2]`. Remember, each hour is 3600 seconds and each minute is 60 seconds. Print the `time_tuple` and its number of seconds since midnight.


````python
# Create the time tuple for 9:16:01 AM
time_tuple = (9, 16, 1)

# Calculate seconds since midnight
seconds_since_midnight = time_tuple[0] * 3600 + time_tuple[1] * 60 + time_tuple[2]

# Print the results
print("Time tuple:", time_tuple)
print("Seconds since midnight:", seconds_since_midnight)
````

### Problem 4b  (1 point)
In the code cell below, write a Python script that unpacks the elements of the `time_tuple` defined above into the variables named `hour`, `minute`, and `second` using one unpacking assignment statement. Repeat the calculation of the number of seconds since midnight, now using the variables you unpacked from the tuple. Print variables' values and the number of seconds since midnight.

````python
# Define the time tuple
time_tuple = (9, 16, 1)

# Unpack the tuple into variables
hour, minute, second = time_tuple

# Calculate seconds since midnight
seconds_since_midnight = hour * 3600 + minute * 60 + second

# Print the results
print("Hour:", hour)
print("Minute:", minute)
print("Second:", second)
print("Seconds since midnight:", seconds_since_midnight)
````
