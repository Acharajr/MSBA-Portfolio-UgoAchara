### Array indexing

```python
monthly_sales = np.array([100, 150, 200, 175, 300, 250])

# 3.1 Single values (like Excel cell references)
print("\nFirst month sales:", monthly_sales[0])  # Python starts at 0!
print("Last month sales:", monthly_sales[-1])  # -1 means last item

# 3.2 Slicing - getting ranges (like selecting Excel ranges)
print("First 3 months:", monthly_sales[0:3])  # Index 0, 1, 2 (3 is excluded)
print("Last 3 months:", monthly_sales[-3:])  # Last three items

revenue_data = np.array([
    [100, 150, 200],  # Q1: Jan, Feb, Mar
    [120, 160, 210],  # Q2: Apr, May, Jun
    [130, 170, 220]   # Q3: Jul, Aug, Sep
])
# 3.3 2D Array indexing (row, column) - like Excel's (A1, B2, etc.)
print("\nQ1, Feb revenue:", revenue_data[0, 1])  # Row 0 (Q1), Column 1 (Feb)
print("All Q1 data:", revenue_data[0, :])  # Row 0, all columns
print("All January data:", revenue_data[:, 0])  # All rows, column 0

# 3.4 Boolean indexing - VERY POWERFUL for filtering data
high_sales = monthly_sales[monthly_sales > 180]  # Get only sales > 180
print("Months with sales > 180:", high_sales)
# will discuss more on this later
```

### Array manipulation and vectorization

```python
prices = np.array([10, 15, 20, 25, 30])
quantities = np.array([5, 3, 2, 4, 6])

# 4.1 Element-wise operations (happens to each item automatically)
# In Excel, you'd need to drag formulas down. NumPy does it automatically!

total_revenue = prices * quantities  # Multiplies each price by its quantity
print("\nRevenue per product:", total_revenue)

# Adding a constant (like a price increase)
#vector operation
new_prices = prices + 2  # Add $2 to each price
print("New prices:", new_prices)

# Percentage calculations
price_increase = prices * 0.10  # 10% increase
print("Price increase amount:", price_increase)
```

```python

# Operations on 2D arrays (operating on rows/columns)


# axis=0 means down the columns 
# axis=1 means across the rows 
A = np.array([[10, 12],
              [15,  9]])  # unit sales (stores x products)

print("Matrix A:")
print(A)
print("Total units:", A.sum())

#important take away from class
print("Col totals (per product):", A.sum(axis=0))
print("Row totals (per store):", A.sum(axis=1))

```

```python
revenue_data = np.array([
    [100, 150, 200],  # Q1: Jan, Feb, Mar
    [120, 160, 210],  # Q2: Apr, May, Jun
    [130, 170, 220]   # Q3: Jul, Aug, Sep
])
print("\n--- 2D Array Operations ---")
print("Revenue data:\n", revenue_data)
print("Total per month (sum down):", np.sum(revenue_data, axis=0))
print("Total per quarter (sum across):", np.sum(revenue_data, axis=1))
print("Average per month:", np.mean(revenue_data, axis=0))
```

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(arr)
# Scalar broadcasting
print("arr + 1:\n", arr + 1)

# Row vector (shape (3,)) adds to each row
row = np.array([10, 10, 10])  # shape (3,)
print("row shape:", row.shape)
print("arr + row:\n", arr + row)

# Column vector (shape (2,1)) adds to each column group by row
col = np.array([[100],
                [200]])       # shape (2,1)
print("col shape:", col.shape)
print("arr + col:\n", arr + col)
```

```python
# Aggregate functions - Getting summary statistics
monthly_sales = np.array([100, 150, 200, 175, 300, 250])

print("\n--- Sales Summary Statistics ---")
print("Total sales:", np.sum(monthly_sales))  # Like Excel SUM()
print("Average sales:", np.mean(monthly_sales))  # Like Excel AVERAGE()
print("Median sales:", np.median(monthly_sales))  # Middle value
print("Min sales:", np.min(monthly_sales))  # Like Excel MIN()
print("Max sales:", np.max(monthly_sales))  # Like Excel MAX()
print("Std deviation:", np.std(monthly_sales))  # Measure of spread

# 4.3 Cumulative operations (running totals)
cumulative_sales = np.cumsum(monthly_sales)  # Running total
print("Cumulative sales:", cumulative_sales)


```

#### Reshaping

- Transpose (switch subscripts: number of rows become number of colume and vice versa)
- Reshape

```python
# Transpose
bdex3=np.array([[4],[3],[2],[1],[23],[-45]])
print(bdex3)
print(bdex3.shape)
print(bdex3.T)
print(bdex3.T.shape)

# Reshape
print(bdex3.reshape(2,3))


# Flatten 
bdex31=np.array([[1,2],[3,4]])
print(bdex31)
print(bdex31.flatten())
```

```python
#Stacking arrays - Combining data from different sources
q1_sales = np.array([100, 150, 200])
q2_sales = np.array([120, 160, 210])
q3_sales = np.array([130, 170, 220])

# Stack vertically (add rows) - like appending data in Excel
all_quarters = np.vstack([q1_sales, q2_sales, q3_sales])
print("\nAll quarters (vertical stack):\n", all_quarters)

# Stack horizontally (add columns) - like adding new columns
product_a = np.array([100, 150, 200])
product_b = np.array([80, 120, 180])
both_products = np.column_stack([product_a, product_b])
print("Both products (horizontal):\n", both_products)


# 6.3 Concatenate - Another way to combine arrays
jan_sales = np.array([100, 120, 140])
feb_sales = np.array([110, 130, 150])
jan_feb = np.concatenate([jan_sales, feb_sales])
print("Jan+Feb sales:", jan_feb)
```

## Boolean and Fancy Indexing

```python


a = np.array([[1,2], [3, 4], [5, 6]])
print(a)

# same shape with True or False for the condition
bool_idx = a > 2
print(bool_idx)

print(a[bool_idx]) # note: this will be a rank 1 array! (returns elements of the matrix for which index is True)


```

```python
# We can do all of the above in a single concise statement: arrayname[question]
print(a[a > 2]) # however, still returns a 1-D array

# try to do it multiple conditions

print(a[(a<3) | (a==3)])

```

```python
sales = np.array([120, 80, 150, 95, 220, 60, 175])
print(np.where(sales > 100))
idx = np.where(sales > 100)[0]
print("Indices where sales > 100:", idx)
```

```python
# boolen indexing
# filter and transform

a = np.array([[1,2], [3, 4], [5, 6]])
print(a)

# np.where(): same size with modified values
b = np.where(a>2, a, 0) # if you want to replace something else except 0, you can do so

# this isused to filter based on a condition 'a>2' if true then 'a' and if false '0' in this example
print(b)

b = np.where(a>2, a, -2) # if you want to replace something else except 0, you can do so
print(b)
```

```python
# fancy indexing: access multiple indices at once
a = np.array([10,19,30,41,50,61])
print(a)
b = a[[1,3,5]]
print(b)


A = np.arange(1, 13).reshape(3, 4)
print("A:\n", A)

rows = np.array([0, 2])   # pick 1st and 3rd rows
cols = np.array([1, 3])   # pick 2nd and 4th cols

# Cross-select (outer product of rows and cols)
print("Selected elements:")
# np.ix_ constructs the correct index arrays for cross-selection
result = A[np.ix_(rows, cols)]

print(result)
```

#### Series
- Data structure with an index and one column

- Aligns on index
- Useful since when accessing a column from a DataFrame, it acts as a Series


```python
import numpy
print(numpy.__version__)

```

```python
import pandas as pd
```

```python

# Defining Series
s = pd.Series([1,2,3,4], index=['a','b','c','d'])
print(s)
# Accessing data from Series
print(s['d'])

#Math with series - aligns on index
s2=pd.Series([10,-20,30,40], index=['a','b','d','c'])
print(s2)
```

```python
print(s2.index)
print(s2.values)
#supports string indexing
print(s2['b'])
# supports boolean indexing
print(s2[s2>0])
```

```python
# label and integer based indexing


s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(s)
# Access a single value by label
print(s.loc['b'])


# Access multiple values by labels
print(s.loc[['a', 'c','d']])

# Access values using a boolean array
print(s.loc[s > 25])

# Access a single value by position
print(s.iloc[1])

# Access multiple values by positions
print(s.iloc[[0, 2]])


# Access a slice of the Series
print(s.iloc[1:3])




```

```python

s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

# Slice from 'b' to 'd' (inclusive)
print(s.loc['b':'d'])

# Slice with specific labels
print(s.loc[['a', 'c', 'e']])

# Slice using a condition
print(s.loc[(s >= 20) & (s <= 40)])

# Slice from position 1 to 3 (exclusive of 3)
print(s.iloc[1:3])

# Slice with a step of 2
print(s.iloc[::2])
```

```python
#  .reindex, .set_index, dropping, slicing, sort index, sort values#

import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 22, 28],
        'City': ['New York', 'London', 'Paris', 'Tokyo']}
df = pd.DataFrame(data)
print(df.head(3))







# .reindex: Rearrange the DataFrame with a new index
new_index = ['a', 'b', 'c', 'd']
df.index= new_index
print(df.head(3))
#print(df)
df_reindexed = df.reindex(['d','b','c','a'])
print("DataFrame after reindexing:")
print(df_reindexed)


# .set_index: Set a column as the index
df_with_name_index = df.set_index('Name')
print("\nDataFrame with 'Name' as index:")
print(df_with_name_index)

# Dropping rows or columns
df_dropped_city = df.drop('City', axis=1)  # Drop the 'City' column
print("\nDataFrame after dropping 'City' column:")
print(df_dropped_city)

df_dropped_row_alice = df.drop('a')  # Drop the row with index a (Alice)
print("\nDataFrame after dropping the first row (Alice):")
print(df_dropped_row_alice)

# Slicing
print("\nSlicing:")
print(df[0:2])  # Select the first two rows
print(df['Name'])  # Select the 'Name' column
print(df[['Name', 'Age']])  # Select 'Name' and 'Age' columns

# Sort index
df_sorted_index = df.sort_index(ascending=False)
print("\nDataFrame sorted by index (descending):")
print(df_sorted_index)

# Sort values
df_sorted_age = df.sort_values('Age')
print("\nDataFrame sorted by Age (ascending):")
print(df_sorted_age)
```
