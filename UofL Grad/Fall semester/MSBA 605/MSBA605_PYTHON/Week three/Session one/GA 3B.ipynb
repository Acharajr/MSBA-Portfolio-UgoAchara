```python
import pandas as pd
import numpy as np

# Create a sample dataset with missing values, duplicates, and errors
data = {
    'Customer_ID': [101, 102, 103, 104, 105, 106, 106, 107, 108, np.nan],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Frank', 'Grace', 'Hannah', 'Ivy'],
    'Age': [25, 30, np.nan, 22, np.nan, 35, 35, 29, 24, 28],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Berlin', 'Paris', 'Paris', 'London', 'Tokyo', 'Berlin'],
    'Purchase_Amount': [120.5, 340.0, np.nan, 560.75, 450.25, np.nan, 340.0, 230.0, 500.5, 320.25]
}

df = pd.DataFrame(data)
print(df)
```

# Tasks

1- Remove any rows where the 'Customer_ID' is missing, as this is a critical column. Drop any other rows where two or more columns have missing values. (1 point)

2- Fill the missing 'Age' values with the mean age of the customers. Fill missing 'Purchase_Amount' values with the median purchase amount.  (1 point)

3- Identify and remove duplicate rows based on the 'Customer_ID' and 'Name' columns. (1 point)

4- Replace all instances of 'Paris' in the 'City' column with 'FR-Paris'  Any negative or incorrect values for the 'Age' or 'Purchase_Amount' should be replaced with np.nan. (1 point)

5- Set the 'Customer_ID' as the new index. (1 point)

```python
#1
df=df.dropna(subset=['Customer_ID'])
df = df.dropna(thresh=4)
print(df)
```

```python
#2
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Purchase_Amount']=df['Purchase_Amount'].fillna(df['Purchase_Amount'].median())
print(df)
```

```python
#3
print(df['Customer_ID'].unique(), '\n')
print(df['Name'].unique(), '\n')
df=df.drop_duplicates(['Name'])
print(df)
```

```python
#4
df['City']=df['City'].replace('Paris','FR-Paris')
print(df)
# Can't change Age or purchase about becuase it does not have Na 
```

```python
#5
print(df.set_index('Customer_ID'),'\n')
```
