### 4 Drop the columns 'name', 'host_id', 'host_name', 'latitude', 'longitude'  from the dataset using .drop([column names]) command. Print the restructured dataset.

```python
#Airbnb_df.drop(['name', 'host_id', 'host_name', 'latitude', 'longitude'], axis = 1, inplace = True)


Airbnb_clean = Airbnb_df.drop(['name', 'host_id', 'host_name', 'latitude', 'longitude'], axis=1)
print("\nAirbnb Clean:\n")
display(Airbnb_clean.head())
```

### 5) Check to see if there are any duplicate rows. Print the numbers.

### 6) Drop the rows where availability is 0 since that means the listing is no longer active. Print the number of rows and columns in the new dataset.

```python
#Airbnb_df.drop(['availability_365'], inplace=True)
print("Dataset shape:",Airbnb_clean.shape)
Airbnb_df = Airbnb_clean[Airbnb_clean["availability_365"] != 0]

# Print the new dataset
print("\nDataset shape after dropping:",Airbnb_df.shape)
display(Airbnb_df.head())
```

### 7) Check for outliers in "price" column and drop the outliers. [2 points]

```python
plt.boxplot(Airbnb_df['price'])
plt.title("Boxplot of Price")
print("Check for outliers in 'price'\n")
plt.show()


# Handle outliers (basic IQR filter for SALES)
Q1 = Airbnb_df['price'].quantile(0.25)
Q3 = Airbnb_df['price'].quantile(0.75)
IQR = Q3 - Q1
lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
Airbnb_df = Airbnb_df[(Airbnb_df['price'] >= lower) & (Airbnb_df['price'] <= upper)]
print("Rows after outlier removal:", len(Airbnb_df))
```

### 8) For each neighbourhood group, find average price and availability. 

```python
# Group by neighbourhood group and calculate average price & availability
Airbnb_df = Airbnb_df.groupby("neighbourhood_group")[["price", "availability_365"]].mean().reset_index()
Airbnb_df.rename(columns={"price": "avg_price", "availability_365": "avg_availability"}, inplace=True)

print("Average price and availability by neighbourhood group\n")
print(Airbnb_df, "\n")

# Average Price Plot 
plt.figure()
plt.barh(Airbnb_df["neighbourhood_group"], Airbnb_df["avg_price"])
plt.title("Average Price by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price ($)")
plt.xticks(rotation=45)
plt.show()

#  Average Availability Plot 
plt.figure()
plt.barh(Airbnb_df["neighbourhood_group"], Airbnb_df["avg_availability"], color="red")
plt.title("Average Availability by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Availability (days)")
plt.xticks(rotation=45)
plt.show()
```
