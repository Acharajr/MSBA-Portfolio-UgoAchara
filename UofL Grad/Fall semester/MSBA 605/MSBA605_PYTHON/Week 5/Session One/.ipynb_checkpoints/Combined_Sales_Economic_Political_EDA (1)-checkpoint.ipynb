# Combined Exploratory Data Analysis 


```python
# 3) Imports and setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Matplotlib defaults
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['axes.grid'] = True
```

##  Sales Data Analysis — Supermarket Sales

```python
import pandas as pd

# Directly downloadable CSV

sales_df = pd.read_csv("sales_data_sample.xls", encoding="cp1252")



display(sales_df.head())
display(sales_df.info())
display(sales_df.describe(include='all'))

```

### Average sale is around $3500

```python
# 1.1 Inspect basic structure
#print("Dataset shape:", sales_df.shape)
#print("\nColumn names:\n", sales_df.columns.tolist())
#print("\nData types:\n", sales_df.dtypes)
print("\nMissing values per column:\n", sales_df.isna().sum())
```

```python
# Handle missing values
# ADDRESSLINE2, STATE, TERRITORY have many nulls → fill with "Unknown"
cols_to_fill = ['ADDRESSLINE2', 'STATE', 'TERRITORY']
for c in cols_to_fill:
    if c in sales_df.columns:
        sales_df[c] = sales_df[c].fillna('Unknown')

# POSTALCODE has few missing → fill with mode
if sales_df['POSTALCODE'].isna().sum() > 0:
    mode_postal = sales_df['POSTALCODE'].mode()[0]
    sales_df['POSTALCODE'] = sales_df['POSTALCODE'].fillna(mode_postal)
```

```python
#  Ensure numeric and date types are correct
numeric_cols = ['QUANTITYORDERED','PRICEEACH','SALES','MSRP','YEAR_ID','MONTH_ID','QTR_ID']
for col in numeric_cols:
    sales_df[col] = pd.to_numeric(sales_df[col], errors='coerce')

# Convert ORDERDATE to datetime if needed
if sales_df['ORDERDATE'].dtype.kind != 'M':
    sales_df['ORDERDATE'] = pd.to_datetime(sales_df['ORDERDATE'], errors='coerce')
```

```python
# Remove duplicates (same order line may be appearing twice)
before = len(sales_df)
sales_df.drop_duplicates(inplace=True)
after = len(sales_df)
print(f"Removed {before - after} duplicate rows")
```

```python
# Handle outliers (basic IQR filter for SALES)
Q1 = sales_df['SALES'].quantile(0.25)
Q3 = sales_df['SALES'].quantile(0.75)
IQR = Q3 - Q1
lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
sales_df = sales_df[(sales_df['SALES'] >= lower) & (sales_df['SALES'] <= upper)]
print("Rows after outlier removal:", len(sales_df))
```

```python
# data wrangling (creating new columns)
sales_df['YEAR_MONTH'] = sales_df['ORDERDATE'].dt.to_period('M').astype(str)
sales_df['YEAR'] = sales_df['ORDERDATE'].dt.year
sales_df['MONTH'] = sales_df['ORDERDATE'].dt.month
```

```python
# 1.7 Verify clean dataset
print("\nAfter cleaning:")
print(sales_df.info())
print("Nulls after cleaning:\n", sales_df.isna().sum())
sales_df.head()
```

**Which product lines drive revenue?**

```python
by_pl = (sales_df.groupby("PRODUCTLINE", as_index=False)["SALES"]
         .sum().sort_values("SALES"))
print(by_pl,'\n')

plt.figure()
plt.barh(by_pl["PRODUCTLINE"], by_pl["SALES"])
plt.title("Total Sales by Product Line")
plt.xlabel("Sales")
#plt.gca().invert_yaxis() #(in case you want the lowest item first)
plt.show()

```




Classic Cars and Vintage Cars sell the most, far more than Trains, Ships, and Planes.

The company earns most of its money from these car models. This shows that customers are very interested in cars, or that the cars may have higher prices.

Because these cars are so popular, the company should focus on them. For example, they could advertise them more or make sure they always have enough in stock.




## Which countries contribute most to sales?

```python
by_country = (sales_df.groupby("COUNTRY", as_index=False)["SALES"]
              .sum().sort_values("SALES"))
print(by_country.head(15),'\n')

plt.figure()
topN = by_country.head(15)
plt.barh(topN["COUNTRY"], topN["SALES"])
plt.title("Top Countries by Sales")
plt.xlabel("Sales")
#plt.gca().invert_yaxis()
plt.show()
```

The United Kigdom contributes the largest sales volume, followed by countries in Europe (France, Spain, UK) and some from Asia (Japan, Singapore). Sales are geographically diverse but concentrated in high-income regions. Evaluate logistics and local marketing spend.

## Monthly sales trend (overall and by year)

```python
monthly = (sales_df.groupby("YEAR_MONTH", as_index=False)["SALES"].sum())
monthly_sorted = monthly.sort_values("YEAR_MONTH")

plt.figure()
plt.plot(monthly_sorted["YEAR_MONTH"], monthly_sorted["SALES"], marker="o")
plt.title("Monthly Sales Trend (All Years)")
plt.xlabel("Year-Month")
plt.ylabel("Sales")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


```

```python
# Create a pivot table

pivot = sales_df.pivot_table(
    index="MONTH",    # Rows will be Months (1-12)
    columns="YEAR",   # Columns will be Years
    values="SALES",   # The cells will be the sum of SALES
    aggfunc="sum",
    fill_value=0      # Automatically fills missing months with 0
)

print(pivot,'\n')
# 2. Plot the pivot table
#    Pandas' .plot() function is smart and will plot each
#    column (each year) as a separate line.
pivot.plot(marker="o", figsize=(10, 6)) # figsize is optional, just makes it bigger

# 3. Add titles and show the plot
plt.title("Monthly Sales by Year")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(range(1, 13)) # Optional: Ensures all 12 month ticks are shown
plt.legend(title="Year")
plt.show()
```

Sales show cyclic peaks toward the end of each fiscal year (around November–December). Seasonality exists (possibly driven by year end gift purchases or corporate orders). Inventory and staffing should ramp up before Q4 and may also analyze lead-times for order fulfillment.

## Order status mix (pipeline health)

```python
status_sales = sales_df.groupby("STATUS")["SALES"].sum().sort_values(ascending=False)
print(status_sales,'\n')

plt.figure()
plt.bar(status_sales.index.astype(str), status_sales.values)
plt.title("Sales by Order Status")
plt.xlabel("Status")
plt.ylabel("Sales")
plt.show()
```

The majority of transactions fall under Shipped or Resolved, but a small fraction are Cancelled or On Hold. The order to delivery pipeline works efficiently overall; however, monitoring the causes of “On Hold” or “Cancelled” could uncover process bottlenecks.

## Top customers (who to prioritize)

```python
by_cust = (sales_df.groupby("CUSTOMERNAME", as_index=False)["SALES"]
           .sum().sort_values("SALES", ascending=False))
print(by_cust.head(20),'\n')

plt.figure()
topC = by_cust.head(15)
plt.barh(topC["CUSTOMERNAME"], topC["SALES"])
plt.title("Top 15 Customers by Sales")
plt.xlabel("Sales")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

A few large corporate customers (e.g., Euro+ Shopping Channel, Mini Gifts Distributor Ltd.) account for a disproportionate share of revenue. Revenue concentration risk exists: losing a top account could materially affect sales. May introduce loyalty or partnership programs for key accounts, while diversifying into mid tier clients.

## Unit price & quantity by product line

```python
agg_pl = (sales_df.groupby("PRODUCTLINE")[["PRICEEACH","QUANTITYORDERED"]]
          .agg(["mean","median"]).round(2))
display(agg_pl)


avg_price_pl = (sales_df.groupby("PRODUCTLINE")["PRICEEACH"].mean()
                .sort_values(ascending=False))
plt.figure()
plt.barh(avg_price_pl.index, avg_price_pl.values)
plt.title("Average Unit Price by Product Line")
plt.xlabel("Average PRICEEACH")
plt.gca().invert_yaxis()
plt.show()

avg_qty_pl = (sales_df.groupby("PRODUCTLINE")["QUANTITYORDERED"].mean()
              .sort_values(ascending=False))
plt.figure()
plt.barh(avg_qty_pl.index, avg_qty_pl.values)
plt.title("Average Quantity Ordered by Product Line")
plt.xlabel("Average QUANTITYORDERED")
plt.gca().invert_yaxis()
plt.show()
```

Classic Cars, Trucks, Buses have the highest unit prices; Trucks & Buses show high average quantity ordered.

```python
import seaborn as sns

# 1. Create the same pivot table
pivot = (sales_df.pivot_table(index="PRODUCTLINE",
                             columns="MONTH",
                             values="SALES",
                             aggfunc="sum").fillna(0))

# 2. Plot using Seaborn
plt.figure(figsize=(12, 7)) # Optional: make it bigger
sns.heatmap(pivot, annot=True, fmt=".0f") # 'annot=True' writes the sales number in each box
plt.title("Seasonality: Sales by Product Line and Month")
plt.show()
```

Clear monthly variation: car-related lines peak around months 5–6 and again at year-end.
Seasonality is product-specific; toy cars might align with holiday or school-break cycles.
