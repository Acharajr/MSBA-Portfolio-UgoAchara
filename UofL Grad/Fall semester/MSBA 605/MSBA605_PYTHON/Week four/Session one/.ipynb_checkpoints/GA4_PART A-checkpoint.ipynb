## 1. Using the dataset, calculate total spending and average transaction per customer across all years. [1 POINT]

```python
import pandas as pd
df.head()
```

```python
df=pd.read_excel(r"CustomerBehavior2.xlsx")

spending = (df.groupby("Name")
                   .agg(Total_spending=("Spent","sum"),
                        Avg_Transaction= ("Transaction","mean")))
print(spending)
```

## 2. Analyze the total Spent for each Customer and Quarter. Replace 0 for missing quarters and add a total column for each customer. [1 point]

```python
spending = pd.pivot_table(
    df,
    values="Spent",
    index = "Name",
    columns = "Quarter",
    fill_value= 0,
    aggfunc= "sum",
    margins=True, 
    margins_name="Total"
)
print(spending)
```

## 3. Analyze the count of how many transactions occurred per Year and Quarter. Show each quarter’s percentage contribution to that year’s total transactions. [1.5 point]

```python
xt = pd.crosstab(df["Year"], 
                df["Quarter"],
                margins= True,
                margins_name="Total")
print(xt,'\n')
xt_prop = pd.crosstab(df["Year"], 
                df["Quarter"],
                      margins= True,
                      margins_name="Total",
                      normalize="columns").round(2)

print(xt_prop)
```

## 4. Generate a grouped summary showing total Spent by each Customer over the Years. Then display the years as columns and customers as rows. Finally, bring it back to its original long form. [1.5 point]

```python
by_region_product = df.groupby(["Year","Name"])["Spent"].sum()
print(by_region_product)

print("\nAs flat columns:")
print(by_region_product.reset_index())
```
