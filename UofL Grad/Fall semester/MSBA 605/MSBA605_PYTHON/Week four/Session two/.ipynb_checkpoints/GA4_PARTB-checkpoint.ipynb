```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent style for the plots
sns.set_style("whitegrid")
```

```python
np.random.seed(101)

# Create a dataset with 20 rows, 3 numerical cols, and 3 categorical cols
data = {
    'Date': pd.to_datetime(np.random.choice(pd.to_datetime(['2025-11-01', '2025-11-02', '2025-11-03', '2025-11-04']), 20)),
    'Item': np.random.choice(['Coffee', 'Tea', 'Pastry', 'Sandwich'], 20),
    'Weather': np.random.choice(['Sunny', 'Cloudy', 'Rainy'], 20),
    'UnitsSold': np.random.randint(5, 50, 20),
    'AvgPrice': np.round(np.random.uniform(3.50, 12.99, 20), 2),
    'CustomerRating': np.random.randint(1, 6, 20) # Ratings from 1 to 5
}

cafe_df = pd.DataFrame(data)

# Create a 'TotalRevenue' column for analysis
cafe_df['TotalRevenue'] = cafe_df['UnitsSold'] * cafe_df['AvgPrice']

print("--- Cafe Sales Dataset ---")
print(cafe_df)
```

```python
cafe_df.info()
```

## 1. Which item category generates the most total revenue? [1 point]

```python
sns.countplot(data=cafe_df, x='Item')
plt.title('Items Ordered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2. What is the proportion of units sold by weather condition? Create a pie chart to show the percentage of total units sold in Sunny, Cloudy, and Rainy weather. [1 point]

```python
Units_sold = cafe_df.groupby('Weather')['UnitsSold'].sum().sort_values(ascending=False)
display(Units_sold)

plt.pie(Units_sold, labels=Units_sold.index, autopct='%1.1f%%', startangle=120)
plt.title('Unit sold  by Weather Condition (Pie)')
plt.show()
```

## 3. How did the total revenue trend change daily? [1 point]

```python
daily = cafe_df.groupby('Date', as_index=False)['TotalRevenue'].sum().sort_values('Date')
sns.lineplot(data=daily, 
             x='Date', 
             y='TotalRevenue', 
             marker='o')
plt.title('Daily Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 4. Is there a relationship between the average price of an item and the number of units sold? [1 point]

```python
sns.scatterplot(data=cafe_df, 
                x='AvgPrice', 
                y='UnitsSold',
               )
plt.title('AvgPrice vs Units Sold')
plt.xlabel('Average price')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()
```

## 5. How does the distribution of customer ratings differ for each item? [1 point]

```python
sns.boxplot(data=cafe_df, 
            x='Item', 
            y='CustomerRating'
           )
plt.title('Customer rating by Item')
plt.tight_layout()
plt.show()
```
