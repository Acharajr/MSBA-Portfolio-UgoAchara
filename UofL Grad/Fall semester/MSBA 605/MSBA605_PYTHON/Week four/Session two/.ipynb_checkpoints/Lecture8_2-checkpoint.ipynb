```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(context="notebook", style="whitegrid")
pd.set_option('display.float_format', lambda x: f"{x:,.2f}")
```

```python
orders_df=pd.read_csv('orders1.csv')   
orders_df.head(5)
```

```python
orders_df.info()
```

```python
customers_df=pd.read_csv('customers1.csv')   
customers_df.head(5)
```

```python
customers_df.info()
```

```python
# combining the datasets and cleaning a bit

merged = pd.merge(orders_df, customers_df, on="CustomerID", how="left")

merged['Rating'] = merged['Rating'].fillna(merged['Rating'].median())
merged['ProductCategory'] = merged['ProductCategory'].str.title().str.strip()
merged['Region'] = merged['Region'].str.title().str.strip()

print("Merged shape:", merged.shape)
merged.head(5)
```

##  Bar Charts
### Simple Bar 

```python
region_rev = merged.groupby('Region', as_index=False)['Total'].sum().sort_values('Total', ascending=False)
display(region_rev)

plt.figure(figsize=(8,5))
sns.barplot(data=region_rev, x='Region', y='Total')
plt.title('Total Revenue by Region')
plt.ylabel('Revenue ($)')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```

###  Stacked Bar 

```python
pivot_region_cat = pd.pivot_table(merged, values='Total', index='Region', columns='ProductCategory', aggfunc='sum', fill_value=0)
display(pivot_region_cat)

ax = pivot_region_cat.plot(kind='bar', stacked=True, figsize=(9,6))
ax.set_title('Revenue by Category within Region (Stacked)')
ax.set_ylabel('Revenue ($)')
ax.set_xlabel('Region')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Grouped (Side-by-Side) Bar 

```python
plt.figure(figsize=(9,6))
sns.barplot(data=merged, x='ProductCategory', y='Total', hue='Region', estimator=np.mean)
plt.title('Average Order Value by Category  Region (Grouped)')
plt.ylabel('Average Order Value ($)')
plt.xlabel('Product Category')
plt.tight_layout()
plt.show()
```

## Pie & Donut Charts 

```python
cat_share = merged.groupby('ProductCategory')['Total'].sum().sort_values(ascending=False)
display(cat_share)

# Pie
plt.figure(figsize=(6,6))
plt.pie(cat_share, labels=cat_share.index, autopct='%1.1f%%', startangle=120)
plt.title('Revenue Share by Category (Pie)')
plt.show()

# Donut
fig, ax = plt.subplots(figsize=(6,6))
wedges, texts, autotexts = ax.pie(cat_share, labels=cat_share.index, autopct='%1.1f%%', startangle=120)
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)
ax.set_title('Revenue Share by Category (Donut)')
plt.tight_layout()
plt.show()
```

## Distributions  Histogram, Box, Violin

```python
# Histogram
plt.figure(figsize=(8,5))
sns.histplot(data=merged, x='Total', bins=20, kde=True)
plt.title('Order Total Distribution (Histogram)')
plt.xlabel('Order Total ($)')
plt.tight_layout()
plt.show()

# Box by Category
plt.figure(figsize=(9,6))
sns.boxplot(data=merged, x='ProductCategory', y='Total')
plt.title('Order Total by Category (Box Plot)')
plt.xlabel('Product Category')
plt.tight_layout()
plt.show()

# Violin by Region
plt.figure(figsize=(9,6))
sns.violinplot(data=merged, x='Region', y='Total', inner='quartile')
plt.title('Order Total by Region (Violin Plot)')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```

## Line & Scatter

```python
# Line: daily revenue trend
daily = merged.groupby('Date', as_index=False)['Total'].sum().sort_values('Date')
plt.figure(figsize=(9,5))
sns.lineplot(data=daily, x='Date', y='Total', marker='o')
plt.title('Daily Revenue Trend (Line)')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter: Rating vs Order Total (colored by Segment)
plt.figure(figsize=(8,5))
sns.scatterplot(data=merged, x='Rating', y='Total', hue='Segment', alpha=0.8)
plt.title('Rating vs Order Total (Scatter)')
plt.xlabel('Customer Rating')
plt.ylabel('Order Total ($)')
plt.tight_layout()
plt.show()
```

## Heatmap  Correlation of Numeric Columns

```python
num_cols = ['Quantity','Price','Rating','Total']
corr = merged[num_cols].corr()
display(corr)

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## Visualizing Pivot-Table / GroupBy Results

```python
# Pivot: Average Quantity by Region  Category
qty_pivot = pd.pivot_table(merged, values='Quantity', index='Region', columns='ProductCategory', aggfunc='mean', fill_value=0)
display(qty_pivot)

plt.figure(figsize=(9,6))
sns.heatmap(qty_pivot, annot=True, cmap='viridis', fmt='.2f')
plt.title('Avg Quantity by Region  Category (Pivot Heatmap)')
plt.tight_layout()
plt.show()

# GroupBy: Top 10 customers by revenue (bar)
top_customers = merged.groupby('CustomerID', as_index=False)['Total'].sum().sort_values('Total', ascending=False).head(10)
display(top_customers)

plt.figure(figsize=(9,5))
sns.barplot(data=top_customers, x='CustomerID', y='Total')
plt.title('Top 10 Customers by Revenue (GroupBy Bar)')
plt.ylabel('Revenue ($)')
plt.xlabel('CustomerID')
plt.tight_layout()
plt.show()
```

## Crosstab Visualizations

```python
# Crosstab: frequency of orders by Region  Category
ct = pd.crosstab(merged['Region'], merged['ProductCategory'])
display(ct)

# Heatmap for counts
plt.figure(figsize=(8,6))
sns.heatmap(ct, annot=True, cmap='Blues', fmt='d')
plt.title('Order Counts by Region  Category (Crosstab Heatmap)')
plt.tight_layout()
plt.show()

# Stacked bar for crosstab
ct.plot(kind='bar', stacked=True, figsize=(9,6), title='Order Counts by Region  Category (Stacked Bar)')
plt.xlabel('Region')
plt.ylabel('Order Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
