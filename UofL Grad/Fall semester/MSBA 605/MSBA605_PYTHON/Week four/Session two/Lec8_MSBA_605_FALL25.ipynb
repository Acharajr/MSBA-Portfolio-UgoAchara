### Importance of formatting
When formatting your charts, decisions should be made based on helping to convey meaning and make them more aesthetically pleasing
- **Color** should be used intentionally (typically to highlight different categories)
    - using color poorly can be confusing and obscure insights
    - color should be accessible to all, including individuals who are color blind (in seaborn sns.set_palette("colorblind"))
    - if your chart will be printed in black and white, make sure your colors are distinguishable in grey scale (I will copy/paste into powerpoint and print preview in grey scale)
- **Legends** should be used to define what colors, linestyles, markers, etc. mean
- **Grid lines** can be used to communicate thresholds but overuse may make charts look busy
- **Axes** must be formatted to be easy to read and have regular intervals
    - y-axis should always start at 0 to facilitate accurate comparisons
    - charts to be compared should have the same axes
    - include axis labels (and sometimes definitions) for clarity
- **Chart titles** can be used to provide clarity to graphs. Two approaches depending on need:
    - tell a story (e.g., 'Website Traffic Growing Exponentially')
    - provide data context (e.g., 'Website Traffic per Quarter')
- **Annotations** can be used to provide a variety of different types of context:
    - display exact values (often requested, may add noise to your graph)
    - if presented in isolation: when was the data pulled, explain odd values, methodology for data collections, data definitions, decisions made during data cleaning, etc.

- **Sorting axes** can facilitate easier understanding of data
    - ***alphabetically*** allows viewer to quickly look up category
    - ***ascending or descending*** can emphasize order of categories
- **Conditional Formatting** can highlight values that exceed a threshold
- **Reference lines** can show your targets or other important data points (often the mean)
- **Highlighting Trends** can uncover patterns in your data (NOTE: do not assume trends, use tools to display true trends)
- **Project forecasts** will clearly display what is a projection versus what is actual data. Projections are helpful but need to be contextualized.

# Line Plot
Use When: You want to show trends over time or continuous data.
Example: Stock prices over time, temperature changes throughout the day, or any data where you need to illustrate the relationship between two continuous variables.
Features: Connects individual data points with lines, making it easy to see trends and changes.

# Bar Plot
Use When: You want to compare quantities across different categories.
Example: Sales figures for different products, survey responses, or frequency counts of categories.
Features: Bars can be vertical or horizontal. Useful for discrete data and can easily show comparisons.

# Stacked Bar Plot
Use When: You want to compare the total values across categories while also showing the composition of those totals.
Example: Sales figures broken down by product for different regions, or demographic breakdowns in survey results.
Features: Bars are stacked on top of each other to show how different categories contribute to the total.

# Histogram
Use When: You want to visualize the distribution of a continuous variable.
Example: Exam scores, ages of participants, or any numerical data where you want to see frequency distribution.
Features: Displays the number of observations within specified intervals (bins), useful for understanding the underlying frequency distribution.

# Box Plot (Box-and-Whisker Plot)
Use When: You want to summarize and visualize the distribution of a dataset, highlighting the median, quartiles, and potential outliers.
Example: Comparing test scores across different classes, or analyzing the spread of data in different groups.
Features: Shows the central tendency, variability, and outliers effectively.

# Scatter Plot
Use When: You want to show the relationship between two continuous variables.
Example: Correlation between hours studied and exam scores, or the relationship between height and weight.
Features: Each point represents an observation; useful for identifying trends, correlations, and clusters.

# Pie Chart
Use When: You want to show proportions of a whole.
Example: Market share of different companies, budget breakdowns, or survey results where the goal is to emphasize the parts of a whole.
Features: Represents data as slices of a pie, making it easy to visualize relative sizes.

#### Figures and subplots

```python
#use plt.figure() to initialize an empty plot object
#has a number of options to specify plot features
fig=plt.figure(figsize=[10,10])

#to create a plot with subplots,  call the add_subplot() function on your figure object
ax1=fig.add_subplot(2,2,1) #creates a subplot in the first quadrant of a 2 row, 2 column plot object
ax2=fig.add_subplot(2,2,2) #creates a subplot in the second quadrant of a 2 row, 2 column plot object
ax3=fig.add_subplot(2,2,3) #creates a subplot in the third quadrant of a 2 row, 2 column plot object
ax3.plot(np.random.randn(50), color='b', marker='o', linestyle='--') #adds a line plot to the last called plot
ax1.hist(np.random.randn(50),bins=7, color='k', alpha=0.3) #adds a histogram to the plot called ax1
ax2.scatter(np.random.randn(50), np.random.randn(50), color='g') #adds a scatterplot to the plot called ax1
#plt.show()

#To create one plot with matplotlib
#fig1=plt.figure(figsize=[5,5])
#plt.scatter(np.random.randn(50), np.random.randn(50), color='g')
```

```python
#plt.subplots() can create subplots in one line of code
fig,axes=plt.subplots(2,2)#, sharex=True, sharey=True) #can specify if axes should be shared; key for accurate comparisons

#build plots by calling each plot index
#can specify features using color, alpha, marker, linestyle
axes[0,0].hist(np.random.randn(50),bins=15, color='k', alpha=0.3)
axes[0,1].scatter(np.random.randn(50), np.random.randn(50), color='g')
#can create an easy legend by feeding a label when creating the plots, then calling legend on the plot
axes[1,0].plot(np.random.randn(50), 'bo--', label='testlabel')
axes[1,0].plot(np.random.randn(50), color='g', marker='o', linestyle='-', label='testlabel1')
#can adjust the white space inbetween plots
#plt.subplots_adjust(hspace=0, wspace=0)

#setting the features of each plots

#specifies location of xticks; can do the same for yticks
axes[0,0].set_xticks([-3,-2,-1,0,1,2,3])

#specifies labels for xticks, must be same length as xticks; can do the same for yticks
#can rotate labels, change font, and font size
axes[0,0].set_xticklabels(['very low', 'low', 'kind of low', 'zero', 'kind of high', 'high', 'very high'],rotation=45, fontsize='small')

#specifies the label for the yaxis; can do the same for xaxis
axes[0,0].set_ylabel('Frequency')

#sets title for each subplot
axes[0,1].set_title('Scatterplot')

#sets legend based on labels fed during creation
axes[1,0].legend(loc='best')

#sets title for overall chart
fig.suptitle('Overall Title')

#example of .annotation()
#provide x coordinates, and the text in the list;
#can also provide y coordinates at this point; but this example uses .25 as the y coordinate for all annotation
annotatelist=[(0.0, 'This is the beginning'), (.5, 'This is the middle'),(.9, 'This is the end') ]
#loop through your list of annotations and plot on the specified plot
for xpos, label in annotatelist:
    axes[1,1].annotate(label, xy=(xpos, .25))

#can also add text using .text()
#provide x coordinate, y coordinate, text to plot, and other specifications
axes[1,1].text(.4,.6,'Test text', family='monospace', fontsize=16)

#save figure out as a .pdf
plt.savefig(r"examplematplot1.pdf",dpi=400)

```

#### Seaborn
- facilitate plotting directly from DataFrames with more control of plot

```python
import seaborn as sns


#creating a new column by combining Year and Quarter
ALLTransactions['YQ']=ALLTransactions['Year'].astype(str)+"_"+ALLTransactions['Quarter'].astype(str)
print(ALLTransactions.head())

#creating a bar plot using ALLTransaction DataFrame with YQ on the xaxis and Spent on the yaxis
#displays means with a measure of variance as error bars (default 95 CI)
ax12=sns.barplot(data=ALLTransactions,x='YQ',y='Spent',hue='Name')
ax12.bar_label(ax12.containers[0], fontsize=10, fmt='%.1f'); # for labels in seaborn
plt.xticks(rotation=45) #rotating xlabels
plt.show()
```

```python
#Creating histogram pf Spent using ALLTransaction DataFrame with different colors for different Names
#displays frequencies of value occurance
sns.histplot(data=ALLTransactions,x='Spent',hue='Name')
plt.show()
```

```python
#Creating scatterplot with trend line using ALLTransactions DataFrame and numitems and spent as x and y values
sns.regplot(x='numitems',y='Spent',data=ALLTransactions)
plt.show()
```

```python
#Creating all possible scatterplots in ALLTransactions DataFrame, with kernel density plots on the diagonal (Default is histograms)
sns.pairplot(data=ALLTransactions, diag_kind='kde')
plt.show()
```

```python
#Creating boxplots of Spent from ALLTransactions DataFrame; grouped by Name
#boxplots display medians and quartiles, plus outliers
sns.boxplot(y='Spent',data=ALLTransactions,x='Name')
plt.show()
```

```python
#Creating violinplots of Spent from ALLTransactions DataFrame; grouped by Name
#violin plots display mirrors of each groups distribution
sns.violinplot(x='Spent',data=ALLTransactions,y='Name',orient='horizontal')
plt.show()
```

```python
#creates a categorical plot with barplots in each pane with each name as a different column, spent on the y axis, and YQ on the x axis
g=sns.catplot(data=ALLTransactions,x='YQ',y='Spent',col='Name',kind='bar')
#rotates all xticklabels
g.set_xticklabels(rotation=45)
plt.show()
```
