# Q2  (2.5 points)-
## You are building a system to check the status of orders in an online store. The store keeps track of orders in a dictionary where the keys are order IDs (strings), and the values are dictionaries containing the status and price of each order. The status can either be "delivered", "processing", or "cancelled".

Create a dictionary called orders with the following structure:

"order_101": {"status": "delivered", "price": 250}
"order_102": {"status": "processing", "price": 150}
"order_103": {"status": "cancelled", "price": 300}
"order_104": {"status": "delivered", "price": 450}
Ask the user to input an order ID.

Use operator chaining to:

Check if the entered order ID exists in the orders dictionary.
If the order exists, check its status:
If the status is "delivered", print "Order <order_id> has been delivered, total price: $<price>."
If the status is "processing", print "Order <order_id> is still being processed."
If the status is "cancelled", print "Order <order_id> was cancelled."
If the order ID does not exist in the dictionary, print "Order not found."
Write the Python code that implements this functionality.

```python
orders ={
    "order_101": {"status": "delivered", "price": 250},
    "order_102": {"status": "processing", "price": 150},
    "order_103": {"status": "cancelled", "price": 300},
    "order_104": {"status": "delivered", "price": 450}
    }
order_id = input("Order ID (eg. order_101): ")
# Used chatgpt here to solve an error: "string indices must be integers, not 'str'" which was a nesting problem.
# Solution was to realising that order_id is nested in orders and not the other way around
if order_id in orders:
    status = orders[order_id]["status"]
    price = orders[order_id]["price"]

    if status == "delivered":
        print(f"Order has been delivered \nTotal price: ${price}.")
    elif status == "processing":
        print("Order is still being processed.")
    elif status == "cancelled":
        print("Order was cancelled.")
else:
    print("Order not found")
```

### Problem 3 (3 points)
In the code cell below, write a Python script that defines a function named `count_words` that accepts a list of one-word strings as its parameter. The function should use a `for` loop to step through each word in the list and create a dictionary with each unique word as a key and a count of each word's frequency in the list as its associated value. Ignore the case of the letters in each word by using string manipulation methods. Return the word counts dictionary. Test your function with the following list of words: `['Pear', 'applE', 'GRApe', 'banANA', 'grape', 'ORange', 'kIwI', 'pear', 'blueBERRY', 'APPle', 'orANge', 'pEAr', 'GRAPE', 'appLE', 'oranGE']`.

```python
def count_words(words):
    word_counts = {}
    

    for word in words:
        word = word.lower() 
        if word in word_counts:
            word_counts[word] += 1 
        else:
            word_counts[word] = 1   

    return word_counts
    
test = ['Pear', 'applE', 'GRApe', 'banANA', 'grape', 'ORange', 'kIwI', 'pear'
        , 'blueBERRY', 'APPle', 'orANge', 'pEAr', 'GRAPE', 'appLE', 'oranGE']
result = count_words(test)
print(result)


    
```
