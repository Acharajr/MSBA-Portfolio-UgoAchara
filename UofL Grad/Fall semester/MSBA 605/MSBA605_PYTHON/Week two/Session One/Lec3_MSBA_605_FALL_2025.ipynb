The rule for rewards is:  
1. A customer qualifies if their **total purchase > \$500**,  
2. OR if their **average purchase per order > \$120 AND num_orders > 3**.  
3. Otherwise, they are not eligible.  

```python
# Customer data stored in a dictionary of dictionaries
customers = {
    "Alice": {"total_purchase": 450, "num_orders": 5},
    "Bob": {"total_purchase": 600, "num_orders": 2},
    "Charlie": {"total_purchase": 300, "num_orders": 4}
}

# --- Alice ---
total = customers["Alice"]["total_purchase"]
orders = customers["Alice"]["num_orders"]
avg = total / orders

if total > 500:
    print("Alice: Reward Eligible")
elif avg > 120 and orders > 3:
    print("Alice: Reward Eligible")
else:
    print("Alice: Not Eligible")
```

## Each repetition of action is known as pass or iteration

| Feature              | **`while` Loop**                                                                                   | **`for` Loop**                                                                     |
| -------------------- | -------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------- |
| **Purpose**          | Repeats **as long as a condition is True**.                                                        | Repeats **for each item** in a sequence (like list, range, string).                |
| **Typical Use Case** | When the number of iterations is **unknown** in advance (e.g., keep asking for input until valid). | When the number of iterations is **known** or you need to go through a collection. |
| **Syntax**           | `while condition:`<br>¬†¬†¬†¬†`# code block`                                                           | `for variable in sequence:`<br>¬†¬†¬†¬†`# code block`                                  |
| **Control Variable** | Must be **manually updated** inside the loop (to avoid infinite loops).                            | Loop variable is **automatically updated** each iteration.                         |
| **Termination**      | Stops when the condition becomes **False**.                                                        | Stops after the sequence is **fully iterated**.                                    |
| **Risk**             | Can create **infinite loops** if condition never becomes False.                                    | Safe‚Äîterminates automatically after last element.                                  |


```python
# while loop
# 1. Initialize a counter variable.
n = 0
# 2. Set the loop condition. The loop will continue as long as 'i' is less than 3.
while n < 3:
    # 3. This is the "body" of the loop - the code that gets repeated.
    print(n,":Hey")
    # 4. IMPORTANT: Update the counter. If you forget this, you get an infinite loop!
    n += 1
```

```python
# Using a for loop 
# This is an even simpler way to do it for a known number of iterations.

# The loop will run for each item in the list [0, 1, 2].
# In the first pass, i will be 0.
# In the second pass, i will be 1.
# In the third pass, i will be 2.
for i in [0,1,2,]:
    print(i,":Hey")
```

# While Loop
## while loop is also called entry-control loop. Condition is tested at top of loop.
    
* Read the data as long as there is input (from a file, standard input,  etc.).
* Statements within loop can execute zero or more times
* Example: A program‚Äôs input loop that accepts values until user enters a   
  sentinel that terminates the input


```python
# Keep reading user input until sentinel 'stop' is entered
# Initialize a variable to hold the total.
total_sales = 0
# Get the first input from the user before the loop starts.
entry = input("Enter sale amount or type 'stop': ")

# The loop continues as long as the user's entry is not "stop".
# .lower() converts the user's input to lowercase, so "Stop" or "STOP" will also work.
while entry.lower() != "stop":
    # Convert the entry (which is a string) to a number (float) and add it to the total.
    total_sales += float(entry) #total_sales = total sales + float(next entry)
    # Get the next input from the user. This is crucial for updating the loop condition.
    entry = input("Enter next sale amount or 'stop': ")

# After the loop finishes, print the final total, formatted to 2 decimal places.
print(f"Total sales entered: ${total_sales:.2f}")
```

```python
# Count the number of digits in a user input and print them

# Initialize a counter for the number of digits.
count=0
# Get a number from the user and convert it to an integer.
num=int(input('enter a number: '))

# The loop continues as long as the number is greater than 0.
while num > 0:
    # The modulo operator (%) gives the remainder of a division.
    # num % 10 always gives the last digit of the number. (e.g., 2345 % 10 is 5)
    print(num%10)

    # Increment the digit counter.
    count+=1

    # Integer division (//) divides and discards the remainder.
    # num // 10 effectively removes the last digit. (e.g., 2345 // 10 is 234)
    num=num//10

# After the loop finishes (when num becomes 0), print the final count.
print('no of digits=',count)
```

# For loop Basics

## iterating through a container

    for variable in container:       # FOR X IN [LIST]/(TUPLE)/{DICT}
         # Loop body: Sub-statements to execute
         # for each item in the container (## Counter or Filter or Transform)
    Statements to execute after the for loop is complete

```python
# Counter
sales = [120, 150, 170, 200, 180, 160, 190, 210, 230, 250]
total = 0
for x in sales:
    total += x
avg = total / len(sales)
print(f"Total={total}, Avg={avg:.2f}")

```

### üîÅ Going Over Containers in Python

| **Technique** | **Typical Use** | **Example Code** | **What It Returns Each Time** | **When to Use It** |
|----------------|----------------|------------------|-------------------------------|--------------------|
| **`for x in list`** | Iterate directly over a list of values. | `for sale in sales:`<br>&nbsp;&nbsp;&nbsp;&nbsp;`print(sale)` | Each element from the list. | When you just need the values. |
| **`for i, x in enumerate(list)`** | Get both **index and value** while looping through a list. | `for i, sale in enumerate(sales):`<br>&nbsp;&nbsp;&nbsp;&nbsp;`print(i, sale)` | A pair ‚Üí (index, value). | When you want item position + value together. |
| **`for a, b in zip(list1, list2)`** | Iterate over **two (or more) lists in parallel**. | `for prod, sale in zip(products, sales):`<br>&nbsp;&nbsp;&nbsp;&nbsp;`print(prod, sale)` | Pairs from each list. | When combining related lists (e.g., products and sales). |
| **`for key, value in dict.items()`** | Iterate over a **dictionary** by key-value pairs. | `for item, price in prices.items():`<br>&nbsp;&nbsp;&nbsp;&nbsp;`print(item, price)` | A pair ‚Üí (key, value). | When working with mappings (e.g., product ‚Üí price). |
| **`for value in dict.values()`** | Iterate only over dictionary **values**. | `for price in prices.values():`<br>&nbsp;&nbsp;&nbsp;&nbsp;`print(price)` | Each value. | When you only care about values. |
| **`for key in dict.keys()`** | Iterate only over dictionary **keys**. | `for item in prices.keys():`<br>&nbsp;&nbsp;&nbsp;&nbsp;`print(item)` | Each key. | When you only need keys (names, IDs, etc.). |



```python
sales = [120, 150, 170]
products = ["Shoes", "Bags", "Watches"]
prices = {"Shoes": 79.5, "Bags": 120.0, "Watches": 249.0}

# in list
for s in sales:
    print(f"Sale amount: {s}")

# enumerate list
for i, s in enumerate(sales):
    print(f"Day {i+1}: Sales = {s}")

# zip two lists
for p, s in zip(products, sales):
    print(f"{p}: {s} units sold")

# dict items
#.item extracts both key and value from a dictionary 
#specific to dictionary 
for item, price in prices.items():
    print(f"{item} price: ${price:.2f}")
```

# Count-Controlled Loops through range()

<span style='color:red'>**Counting using the range() function**</span>

* range(Y) generates a sequence of all non-negative integers less than Y.
    
    Ex: range(3) creates the sequence 0, 1, 2.
    
* range(X, Y) generates a sequence of all integers >= X and < Y.
    
    Ex: range(-7, -3) creates the sequence -7, -6, -5, -4.
    
* range(X, Y, Z), where Z is positive, generates a sequence of all integers >=  X and < Y, incrementing by Z.

  Ex: range(0, 50, 10) creates the sequence 0, 10, 20, 30, 40.
  
* range(X, Y, Z), where Z is negative, generates a sequence of all integers <= X and > Y, incrementing by Z.

  Ex: range(3, -1, -1) creates the sequence 3, 2, 1, 0.

```python
#range(x) will return a sequence from 0 to x-1
for y in range(5):
    print(y)
```

```python
'''Program that calculates savings and interest'''

initial_savings = int(input('Enter the principal: '))
interest_rate = float(input('Enter your bank interest rate: '))/100

years = int(input('Enter years: '))
print()

savings = initial_savings
for i in range(years):
    print(f' Savings in year {i}: ${savings:.2f}')
    savings = savings + (savings*interest_rate)

print('\n')
```

# Loops with conditional

```python
'''
Program that has a conversation with the user.
Uses elif branching and a random number to mix up the program's responses.
'''
import random  # Import a library to generate random numbers

print('Tell me something about yourself.')
print('You can type \'Goodbye\' at anytime to quit.\n')

user_text = input()

while user_text != 'Goodbye':
    random_num = random.randint(0, 2)  # Gives a random integer between 0 and 2
    if random_num == 0:
        print('\nPlease explain further.\n')
    elif random_num == 1:
        print(f"\nWhy do you say: '{user_text}'?\n")
    elif random_num == 2:
        print('\nWhat else can you share?\n')
    else:
        print('\nUh-oh, something went wrong. Try again.\n')

    user_text = input()

print('It was nice talking with you. Goodbye.\n')
```

# Skipping and stopping in loops

* <span style='color:red'>break:</span>	Terminate the current loop. Use the break statement to come out of the  loop instantly.
* <span style='color:red'>continue:</span>	Skip the current iteration of a loop and move to the next iteration
* <span style='color:red'>pass:</span>	Do nothing. Ignore the condition in which it occurred and proceed to run the program as usual

```python
numbers = [10, 40, 120, 230, 100, 23]
iteration=0
for i in numbers:
    iteration+=1
    if i > 220:
        break
    print(iteration)
    
    print('current number', i)
```

```python
# break outer loop in nested loop
for i in range(1, 11):
    # condition to break outer loop
    if i > 5:
        break
    print('Multiplication table of', i)
    for j in range(1, 11):
        print(i * j, end=' ')
    print('')
```

# continue statement

```python
# continue: skip invalid rows
rows = [{"price":10}, {"price":None}, {"price":25}]
clean = []
for r in rows:
    if r["price"] is None:
        continue
    clean.append(r["price"])
print(clean)
```
