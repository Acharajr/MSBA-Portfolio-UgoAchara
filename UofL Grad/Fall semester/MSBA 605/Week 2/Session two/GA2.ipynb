### Problem 1 (2 points)
In the code cell below, write a Python script that inputs an integer dollar amount (from 1 - 99 for this application) to be withdrawn from our virtual ATM. You may assume that the user follows your directions for the input (namely, entering an integer between 1 and 99 inclusive). Calculate and print the number of bills of each denomination (twenties, tens, fives, and ones) that should be dispensed, minimizing the total number of bills used. Start with twenties and work your way down to ones. If no bills of a particular denomination are needed, skip that bill in the output using `if` statements. (10 pts.)
### Sample Output
<pre>
Welcome to the Virtual ATM!
Enter integer dollar amount (1 - 99): 94
    
Dispense:
4 twenties
1 tens
4 ones
</pre>

**Hint**: No _fives_ are printed in the above example because the entered dollar amount didn't require any. Your code, however, must include logic for every denomination (twenties, tens, fives, and ones). An easy way to approach this problem is with the use of floor division (`//`) and remainder (`%`) operations. In this example, the number of twenties needed is `94 // 20`, which gives 4. The remaining amount to be dispensed then is `94 % 20`, which is 14. The number of tens needed then is `14 // 10`, which gives 1. The remaining amount to be dispensed then is `14 % 10`, which is 4. The number of fives needed then is `4 // 5`, which gives 0 (so skip the output). The remaining amount to be dispensed then is `4 % 5`, which is still 4  but we know that should be dispensed as ones (since it's the lowest denomination left).

````python
print("Welcome to the Virtual ATM!")
amount = int(input("Enter integer dollar amount (1 - 99): "))
print("\nDispense:")
twenties = amount // 20  
amount = amount - (twenties * 20) 
tens = amount // 10
amount = amount - (tens * 10)
fives = amount // 5
amount = amount - (fives * 5)
ones = amount
if twenties > 0:
    print(twenties, "twenties")
if tens > 0:
    print(tens, "tens")
if fives > 0:
    print(fives, "fives")
if ones > 0:
    print(ones, "ones")
````

## Problem 3 (2 points)

You are given a list of sales amounts (in USD): sales = [80, 120, 95, 150, 220, 60, 175]. 
Keep only sales above $100 and apply a 5% tax to each of them. Store the tax-inclusive values in a new list called `with_tax`. Finally, print both the `with_tax` list and the total sales from this list.




````python
sales = [80, 120, 95, 150, 220, 60, 175]
with_tax = []

for sale in sales:
    if sale > 100:
        with_tax.append(sale * 1.05)
        total = sum(with_tax)
print(f"Sale with tax: {with_tax} \nTotal Sale:{total}")
````

## Question 6  (2 ponits)

You are given a list of **raw customer records** containing messy text data.  
Each record has the format:  
`"<Name> , <Email> , <City>"`

1. Create a new list called `cleaned_records`, where each record is a **dictionary** with keys:
   - `"name"` → cleaned and title cased name  
   - `"email"` → lowercase email, stripped of spaces  
   - `"city"` → title cased city name  
2. Extract only the **domain** part (text after `@`) from each email and store it in a new list called `email_domains`.
3. Print both `cleaned_records` and `email_domains`.

Hint: Use **string methods** (`.strip()`, `.lower()`, `.title()`, `.split()`, `.replace()`, etc.) and **list comprehensions** or simple loops.




````python
raw_records = [
    " alice  , ALICE@Example.COM ,  new york ",
    "BOB SMITH,  bob.smith@Gmail.Com,los angeles",
    "   charlie,CHARLIE@Yahoo.com,  chicago"
]

# cleaned_records using a list comprehension (used chatgot to figure how to use the functions to clean the data as well as what .title and .replace does)
cleaned_records = [
    {
        "name": name.strip().title(),
        "email": email.strip().lower(),
        "city": city.strip().title()
    }
    for name, email, city in [customer_info.split(",") for customer_info in raw_records]
]

# Extract email domains using another list comprehension
email_domains = [customer_info["email"].split("@")[1] for customer_info in cleaned_records]

# Print results
print("Cleaned Records:",cleaned_records)
print("\nEmail Domains:",email_domains)

````
